```python
from gtts import gTTS
from IPython.display import Audio

# User input
text = input("Enter text to convert to speech: ")

# Language
language = 'en'

# Generate TTS
tts = gTTS(text=text, lang=language)

# Save as mp3
output_file = "speech.mp3"
tts.save(output_file)

# Play audio in Colab
Audio(output_file)
```
